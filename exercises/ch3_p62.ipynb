Draw the straight line between (2,3) and (6,5). Write its equation inthe form y = ax + b. How much deos this line shift to give the least-squares line that fits to these two points and their point if the third point is: 

a. (4,5) </br>
b. (4,2) </br>
c. (5,5) </br>
d. (3,2)

```python
import numpy as np
from sympy import symbols
from naf import incf
import matplotlib.pyplot as plt
```

```python
z = symbols('z')
pt1 = np.array([2,3],dtype='float')
pt2 = np.array([6,5],dtype='float')

m = (pt2[1] - pt1[1])/(pt2[0] - pt1[0])

y = m*(z-pt1[0]) + pt1[1]
print(y)
```

```python
x1 = np.array([2,6], dtype='float')
y1 = np.array([3,5], dtype='float')

x2 = np.array([4,4,5,3], dtype='float')
y2 = np.array([5,2,5,2], dtype='float')

a = np.zeros((x2.shape[0],2))
i = 0
for x,y in zip(x2, y2):
    x = np.append(x1, x)
    y = np.append(y1, y)
    pts = np.column_stack((x,y))
    a[i] = incf.least_squares_polynomial(pts,1)
    i += 1

print(a)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.scatter(x1,y1)
ax.scatter(x2,y2)
ax.plot(x1,y1)
for n in a:
    x = np.array([2,6],dtype='float')
    y = np.array([n[0]+n[1]*x[0], n[0]+n[1]*x[1]])
    ax.plot(x,y)
plt.show()
```

The first two points, (4,5) and (4,2), shift the line vertially up and down without changing the slope because the x-coordinate of the first two points is the average x-coordinate of the points for the original line. 

The third point, (5,5), shifts the line up slightly and steepens the slope slightly because it is above the original line. There is more of a deviation of the original line near the second original point (6,5) because the the added point (5,5) is closer to this point. 

The fourth point, (3,2), shifts the line down and steepens the slope because the line is below the original line and further from the original line. There is more deviation of the original line near the first original point (2,3) because the added point (3,2) is closer to this point. 

This is quite interesting and somewhat reminds me of Bezier and B-spline curves, changing the behavior of the curve with points that don't lie on the curve.
