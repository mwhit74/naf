Find the determinant of the matrix<br>
by row operations to make it<br>
a. upper triangular<br>
b. lower triangular<br>


```python
import numpy as np
from naf.linalg import gedo, gecr, det
```

```python
a = np.array([[2,5,-1],
              [1,6,4],
              [7,-4,2]], dtype=float)
```

```python
lu, ov = gedo(a, False)
```

```python
print(lu, ov)
```

```python
print(f'Determinant: {det(lu,ov)}')
```

```python
lu1, ov1 = gecr(a, False)
```

```python
print(lu1, ov1)
```

```python
print(f'Determinant: {det(lu1,ov1)}')
```
