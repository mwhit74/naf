Repeat p49, this time for a cubic B-spline curve

```python
import numpy as np
import matplotlib.pyplot as plt
from naf import linalg
```

```python
cbc = np.empty((101,4))

cm = np.array([[-1,3,-3,1],
               [3,-6,3,0],
               [-3,0,3,0],
               [1,4,1,0]])

cm = 1/6*cm

i = 0
for u in np.linspace(0,1,101):
    uv = np.array([u**3,u**2,u,1])
    cbc[i] = linalg.mvm(cm, uv)
    i+=1

x = np.linspace(0,1,101)
fig,ax = plt.subplots(figsize=(10,10))
ax.grid()
ax.plot(x,cbc[...,0],x,cbc[...,1],x,cbc[...,2],x,cbc[...,3])
plt.show()
```
