For f(x) as defined here, find polynomials of degrees 2, 3, 4, and 5 that fit f(x) as equally spaced points in [-1,1]. Plot these values and observe that the fit is poor.

f(x) = </br>
0, -1 < x < -0.25 </br>
1-abs(4x), -0.25 < x < 0.25 </br>
0, 0.25 < x < 1

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
from naf.incf import od_tb, od_poly
from sympy import symbols, simplify
```

```python
def f1(x):
    if x > -1 and x < -0.25:
        return 0
    if x > -0.25 and x < 0.25:
        return 1 - abs(4*x)
    if x > 0.25 and x < 1:
        return 0

v_f1 = np.vectorize(f1, otypes=[float])
x1 = np.linspace(-0.99, 0.99, 51)
y1 = v_f1(x1)

fig, ax = plt.subplots()
ax.grid()
ax.plot(x1, y1)
plt.plot()
```

```python
z = symbols('z')
x2 = np.array([-0.625, 0.0, 0.625])
y2 = v_f1(x2)

tb2 = od_tb(y2)
peq2 = od_poly(tb2[0], x2[0], 0.625, z, 2)
print(simplify(peq2))
eq2 = lambda z: 1.0 - 2.56*z**2
veq2 = np.vectorize(eq2)
y2 = veq2(x1)

fig, ax = plt.subplots()
ax.grid()
ax.plot(x1, y2, x1, y1)
plt.show()
```

```python
x3 = np.array([-0.45, -0.15, 0.15, 0.45])
y3 = v_f1(x3)

tb3 = od_tb(y3)
peq3 = od_poly(tb3[0], x3[0], 0.30, z, 3)
print(simplify(peq3))
eq3 = lambda z: -2.22222222222222*z**2 + 2.22044604925031e-16*z + 0.45


veq3 = np.vectorize(eq3)
y3 = veq3(x1)

fig, ax = plt.subplots()
ax.grid()
ax.plot(x1, y3, x1, y1)
plt.show()
```

```python
x4 = np.array([-0.30, -0.15, 0.0, 0.15, 0.30])
y4 = v_f1(x4)

tb4 = od_tb(y4)
peq4 = od_poly(tb4[0], x4[0], 0.15, z, 4)
print(simplify(peq4))
eq4 = lambda z: 230.452674897119*z**4 - 31.8518518518519*z**2 - 4.44089209850063e-16*z + 1.0


veq4 = np.vectorize(eq4)
y4 = veq4(x1)

fig, axs = plt.subplots(1,2, figsize=(10,5))
axs[0].grid()
axs[0].plot(x1, y4, x1, y1)

axs[1].grid()
axs[1].set_xlim(-0.35, 0.35)
axs[1].set_ylim(-0.5, 1.5)
axs[1].plot(x1, y4, x1, y1)
plt.show()
```

```python
x5 = np.array([-0.30, -0.18, -0.06, 0.06, 0.18, 0.30])
y5 = v_f1(x5)

tb5 = od_tb(y5)
peq5 = od_poly(tb5[0], x5[0], 0.12, z, 5)
print(simplify(peq5))
eq5 = lambda z: 136.638374485597*z**4 - 21.5856481481482*z**2 - 4.44089209850063e-16*z + 0.8359375

veq5 = np.vectorize(eq5)
y5 = veq5(x1)

fig, axs = plt.subplots(1,2, figsize=(10,5))
axs[0].grid()
axs[0].plot(x1, y5, x1, y1)

axs[1].grid()
axs[1].set_xlim(-0.35, 0.35)
axs[1].set_ylim(-0.5, 1.5)
axs[1].plot(x1, y5, x1, y1)
plt.show()
```

Some localized matching can be found but away from the specified points used to define the polynomials the polynomial function has significant error compared to the actual function
