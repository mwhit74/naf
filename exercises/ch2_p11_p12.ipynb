```python
If the r.h.s of p11 is transpose([2,3,1,3]) show that there
are an infinite number of solutions.
"""
```

```python
import numpy as np
from naf.linalg import gega, gesv
```

```python
a = np.array([[3,2,-1,-4],
              [1,-1,3,-1],
              [2,1,-3,0],
              [0,-1,8,-5]],dtype=float)
```

```python
b1 = np.array([10,-4,16,3], dtype=float)
b2 = np.array([2,3,1,3], dtype=float)
```

```python
lu, ov = gega(a)
```

```python
print(lu, ov)
```

```python
print(gesv(lu, ov, b1))
```

```python
print(gesv(lu, ov, b2))
```
