```python
import numpy as np
from naf.linalg import tddo, tdsv
import pdb
```

```python
a = np.array([[0, -4, 2],
              [1, -4, 1],
              [1, -4, 1],
              [1, -4, 1],
              [2, -4, 0]], dtype=float)
```

```python
b = np.transpose(np.array([0,-4,-11,5,6], dtype=float))
```

```python
lu = tddo(a)
```

```python
x = tdsv(lu, b)
```
