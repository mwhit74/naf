```python
import sys
import numpy as np
from naf.linalg import gedo
```

```python
a = np.array([[5, 3, 1],
              [3, -1, 2],
              [-5, -3, -1]], dtype=float)
```

```python
try:
    lu, ov = gedo(a, False)
except Exception as e:
    print(e)
else:
    print(lu, ov)
```
