If the degree of the least-squares polynomial is exactly one less than the number of points, the polynomial passes through all of the points; that is, it is an interpolating polynomial. Fit a fifth-degree polynomial to the six points of p59. Now plot this polynomial and compare it to plots of p59 and the least-squares quadratic of p70. Compare the computed y-values at x = 1.5, 2.5, 3.5, 4.5, and 5.5 from the three relations.

```python
import numpy as np
from naf import incf
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5,6])
y = np.array([5.04,8.12,10.64,13.18,16.20,20.04])

pts = np.column_stack((x,y))

ap5 = incf.least_squares_polynomial(pts, 5)

xp5 = np.linspace(1,6,50)
yp5 = incf.eval_poly(ap5, xp5)
```

```python
ap1 = incf.least_squares_polynomial(pts,1)

xp1 = np.linspace(1,6,50)
yp1 = incf.eval_poly(ap1, xp1)
```

```python
ap2 = incf.least_squares_polynomial(pts,2)

xp2 = np.linspace(1,6,50)
yp2 = incf.eval_poly(ap2, xp2)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.scatter(x,y)
ax.plot(xp1,yp1)
ax.plot(xp2,yp2)
ax.plot(xp5,yp5)
plt.show()
```

Since the data is "experimental" the 5th degree polynomial seems to me to over-fit the data. Because it's "experimental" we want to get the general trend as function not match each point as if it is an exact point, all experiments have some error. 

Quantatively the quadaratic fits the data better than the linear model. However, the curve in the quadratic is so slight that without more data or a theoretical basis on which to support the use of a quadratic it seems more complex than necessary.

```python
x2 = np.array([1.5,2.5,3.5,4.5,5.5])

yp1 = incf.eval_poly(ap1, x2)
yp2 = incf.eval_poly(ap2, x2)
yp5 = incf.eval_poly(ap5, x2)

print(yp1)
print(yp2)
print(yp5)

fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.scatter(x2, yp1)
ax.scatter(x2, yp2)
ax.scatter(x2, yp5)
plt.show()
```

```python
e1 = yp2 - yp1
e2 = yp5 - yp1
ee = e1 - e2
print(e1)
print(e2)
print(ee)
```
