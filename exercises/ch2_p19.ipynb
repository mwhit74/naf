```python
from sympy import symbols, simplify, factor
```

```python
n = symbols('n')
```

```python
f1 = n*(n-1)/2
f2 = n*(2*n-1)*(n-1)/6
f3 = f1
```

```python
f4 = simplify(f1 + f2 + f3)
print(f4)
```
