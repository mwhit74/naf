```python
import numpy as np
import matplotlib.pyplot as plt
from naf.incf import lagrangian_poly as lag_poly
from functools import partial
```

```python
equ = lambda x: 0.617*x**2 - 0.105*x - 1.405
vequ = np.vectorize(equ)

pts = np.array([[-2.3,2.1],[0.5,-1.3],[3.1,4.2]])
n = 2


f1 = lag_poly(pts, n, pts[0,0])
f2 = lag_poly(pts, n, pts[1,0])
f3 = lag_poly(pts, n, pts[2,0])

x = np.linspace(-3, 4)
fequ = vequ(x)

fp = partial(lag_poly,pts,n)
f = list(map(fp, x))


fig, ax = plt.subplots(2,1,figsize=(10,10))
ax[0].scatter(pts[:,0], pts[:,1])
ax[0].plot(x, fequ)
ax[0].grid()
ax[0].plot()

ax[1].scatter(pts[:,0], pts[:,1])
ax[1].plot(x, f, 'r')
ax[1].grid()
ax[1].plot()
```
