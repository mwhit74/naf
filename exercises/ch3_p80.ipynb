Fit f(x) = 2x - 1 + exp(-10x^2) between x = -1 and x = 1 with polynomials of degree 2,3,4,5,and 6 
that match f(x) at equally spaced points. Compare the graphs of these to the plot of f(x).

```python
import numpy as np
from naf.incf import od_tb, od_poly
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
def fp(x):
    return 2*x - 1 + np.exp(-10*x**2)

x = []
y = []
tb = []
for i in range(6):
    x.append(np.linspace(-1,1,i+3))
    y.append(fp(x[i]))
    tb.append(od_tb(y[i]))

xp = np.linspace(-1,1,50)

data = np.empty((6,50))
data[0] = fp(xp)

for i in range(1,6):
    data[i] = od_poly(tb[i][0], x[i][0], x[i][1] - x[i][0], xp, i+2)

fig, ax = plt.subplots(5,1,figsize=(10,20))

for i in range(5):
    ax[i].set_title(str(i+2) + " Degree Polynomial")
    ax[i].plot(xp, data[0])
    ax[i].plot(xp, data[i+1])
    ax[i].grid()
    
plt.show()
```
