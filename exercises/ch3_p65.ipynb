Observe that the following data seem to be fit by a curve y=a*\exp(bx) by plotting on semilog paper and noting that points appear to fall on a straight line. (The data are the solubilites of n-butane in anhydrous hydrofluoric acid at high pressures and were used in the design of petroleum refineries.) Find values of a and b from the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.array([77,100,185,239,285],dtype='float')
s = np.array([0.024,0.034,0.07,0.111,0.196])
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.semilogy(t,s)
plt.show()
```
