```python
import numpy as np
from naf.linalg import gedo, dosv
```

```python
np.set_printoptions(precision=3, suppress=True)
```

```python
a = np.array([[2, 1, 3],
              [-1, 0, 2],
              [6, 2, 2]], dtype = float)
```

```python
b = np.array([1, 0, 0], dtype=float)
```

```python
lu, ov = gedo(a, True)
```

```python
for i in range(len(ov)):
    print(lu[ov[i]])
```

```python
x = dosv(lu, ov, b)
```

```python
print(x)
```
