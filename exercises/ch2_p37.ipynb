a. Find the determinant of the Hilbert matrix in Exercise 31.<br>
A small value of the determinant indicates ill-conditioning<br>
(when the matrix has element on the order of unity).<br>
b. Find the inverse of the Hilbert matrix in Exercise 31.<br>
The inverse of an ill-conditioned matrix has some very<br>
large elements in comparison to the elements of the original<br>
matrix.<br>


```python
import numpy as np
from naf.linalg import gedo, det
```

```python
a = np.array([[1, 1/2, 1/3, 1/4],
              [1/2, 1/3, 1/4, 1/5],
              [1/3, 1/4, 1/5, 1/6],
              [1/4, 1/5, 1/6, 1/7]], dtype=float)
```

```python
lu, ov = gedo(a)
```

```python
print(lu, ov)
```

```python
print(f'Determinant: {det(lu,ov)}')
```
