Which of these matrices are singular?<br>


```python
import sys
import numpy as np
from naf.linalg import gedo
```

```python
a = np.array([[-2, 3, -2],
              [1, -4, 7],
              [-1, -6, 17]], dtype=float)
```

```python
try:
    lu, ov = gedo(a)
except Exception as e:
    print(e)
    sys.exit()
```

```python
print(lu, ov)
```
