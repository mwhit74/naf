It is suspected (from theoretical considerations) that the rate of flow from a fire hose is proportional to some power of the pressure at the nozzel. Do the following data confirm the speculation? What is the least-squares value of the exponent? (Assume the pressure data are more accurate.) 

Note:
See PDF notebook for derivation of equations used below.

```python
import numpy as np
from naf import linalg
import math
import matplotlib.pyplot as plt
```

```python
f = np.array([94,118,147,180,230],dtype='float')
p = np.array([10,16,25,40,60],dtype='float')

n = f.shape[0]

nm = np.array([[n, np.sum(np.log(p))],
              [np.sum(np.log(p)), np.sum(np.log(p)**2)]])

b = np.array([np.sum(np.log(f)), np.sum(np.log(f)*np.log(p))])

ap = linalg.doqsv(nm, b)
print(ap)

a = math.exp(ap[0])
b = ap[1]

print(a,b)

eq = lambda x: a*math.pow(x,b)
eq = np.vectorize(eq)
x = np.linspace(5,75,50)
y = eq(x)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.scatter(p,f)
ax.plot(x,y)
ax.set_yscale('log')
ax.set_xscale('log')
plt.show()
```
