Form the difference table for f(x) = 2x^3 - 4x^2 - 2x + 3 for the interval [-1,1] with a spacing of 0.2. Are the third differences a constant as expected? Is the value of this constant equal to a_0\*n!\*h^n ?

```python
import numpy as np
from naf.incf import od_tb
from tabulate import tabulate
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-1,1,11)

y = lambda x: 2*x**3 - 4*x**2 - 2*x + 3
yv = np.vectorize(y)

fx = yv(x)

pts = np.column_stack((x,fx))

print(pts)
```

```python
tb = od_tb(fx)

tb = np.column_stack((x,tb))
tb_h = ['x', 'f(x)', 'D1f', 'D2f', 'D3f', 'D4f', 'D5f', 'D6f', 'D7f', 'D8f', 'D9f', 'D10f']
print(tabulate(tb, tb_h, floatfmt='.4f'))
```

Part (a)

The third differences are a constant value of 0.096 as expected and all differences after the third difference are zero.

```python
a0 = 2.016
n = 3.0
h = 0.2

nf = math.factorial(n)

hpn = h**n

c = a0*nf*hpn
print(c)
```

Part (b)

The third difference value and a_0\*n!\*h^n are similar. The error is approximately 0.0007.
