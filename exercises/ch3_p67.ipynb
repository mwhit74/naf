Find the least-square values for the parameters of y = a\*exp(bx) by fitting the data of p65 to the relation ln(y) = ln(a) + bx.

Note: See PDF notebook for derivation of simultaneous equations solved below.

```python
import numpy as np
from naf import linalg
import math
import matplotlib.pyplot as plt
```

```python
t = np.array([77,100,185,239,285],dtype='float')
s = np.array([0.024,0.034,0.07,0.111,0.196])
```

```python
nm = np.array([[5, np.sum(t)],
               [np.sum(t), np.sum(t**2)]])
               
b = np.array([np.sum(np.log(s)),np.sum(np.log(s)*t)])

ap = linalg.doqsv(nm, b)
print(ap)

a = math.exp(ap[0])
b = ap[1]
print(a,b)

eq = lambda x: a*math.exp(b*x)
x = np.linspace(50,300)
eq = np.vectorize(eq)
y = eq(x)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.scatter(t,s)
ax.plot(x,y)
ax.set_yscale('log')
plt.show()
```
