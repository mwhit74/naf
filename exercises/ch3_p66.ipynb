Plot the data of p65 on ordinary graph paper, and observe that the data are nonlinear.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.array([77,100,185,239,285],dtype='float')
s = np.array([0.024,0.034,0.07,0.111,0.196])
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.plot(t,s)
plt.show()
```
