```python
import matplotlib.pyplot as plt
import numpy as np
import math
from naf.nonlin import fpi, muller, secant, bisect, newtone
f = lambda z: math.tan(z) - z
f2 = lambda z: math.tan(z)
f3 = lambda z: math.atan(z)
df = lambda z: pow(math.tan(z), 2)
```

```python
def graph(x1, x2, num=100):
    vf = np.vectorize(f)
    x = np.linspace(x1, x2, num=num)
    
    fig, axes = plt.subplots()
    axes.grid()
    axes.plot(x, vf(x))
    plt.show()

graph(-5,5)
graph(-4.72,-4.7, num=1000)
graph(-1.58,-1.56, num=1000)
graph(1.4,1.6, num=1000)
graph(4.6,4.8, num=1000)
```
