The data in p59, although actually perturbations from a straight line, plot better along a curve because of the accidental occurance of three negative deviations in succession. Fit a quadratic to the data. How do the sums of the squares of the deviations compare between the line and the curve?

```python
import numpy as np
from naf import incf
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5,6])
y = np.array([5.04,8.12,10.64,13.18,16.20,20.04])

pts = np.column_stack((x,y))
```

```python
ap1 = incf.least_squares_polynomial(pts,1)

xp1 = np.linspace(1,6,50)
yp1 = incf.eval_poly(ap1, xp1)
```

```python
ap2 = incf.least_squares_polynomial(pts,2)

xp2 = np.linspace(1,6,50)
yp2 = incf.eval_poly(ap2, xp2)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.grid()
ax.scatter(x,y)
ax.plot(xp1,yp1)
ax.plot(xp2,yp2)
plt.show()
```

Linear Equation - Sum of squares of deviations

```python
yep1 = incf.eval_poly(ap1, x)
s1 = np.sum((y - yep1)**2)

print(yep1)
print(s1)
```

Quadratic Equation - Sum of squares of deviations

```python
yep2 = incf.eval_poly(ap2, x)
s2 = np.sum((y - yep2)**2)

print(yep2)
print(s2)
```

As can be seen above the total sum of squares of deviations (SSD) for the linear equation is 0.7834 and for the quadratic equation is 0.3829. 

This means that the quadratic equation is a better fit for the data and the linear equation because it produces less overall error when considering all the data points.
