The Hilbert matrix is a classic case of the pathological situation<br>
called "ill-conditioning". The 4x4 Hilbert matrix is:<br>
For the system Hx=b with b^T = [25/12, 77/60, 57/60, 319/420], the<br>
exact solution is x^T = [1,1,1,1]<br>
a. Show that the matrix is ill-conditioned by showing that it is <br>
nearly singular<br>
b. Using only three significant digits (chopped) in your<br>
arithmetic, find the solution to Hx=b. Explain why the answers<br>
are so poor.<br>
c. Using only three significant digits, but rounding, again<br>
find the solution and compare it to that obtained in part (b).<br>


```python
import numpy as np
from naf.linalg import gedo, dosv, dofe, dobs
```

```python
np.set_printoptions(precision=4, formatter={'float':'{:0.4f}'.format}, suppress=False)
```

```python
H = np.array([[1, 1/2, 1/3, 1/4],
              [1/2, 1/3, 1/4, 1/5],
              [1/3, 1/4, 1/5, 1/6],
              [1/4, 1/5, 1/6, 1/7]], dtype=float)
```

```python
H = np.around(H,3)
```

```python
print(H)
```

```python
lu, ov = gedo(H, False)
```

```python
lu = np.around(lu, 4)
```

```python
b = np.array([25/12, 77/60, 57/60, 319/420], dtype=float)
```

```python
b = np.around(b, 3)
```

```python
print(b)
```

```python
print(lu, ov)
```

```python
c = dofe(lu, ov, b)
```

```python
print(c)
```

```python
x = dobs(lu, ov, c)
```

```python
print(x)
```
