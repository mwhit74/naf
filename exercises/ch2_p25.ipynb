```python
from naf.linalg import gedo, dosv, gecr, crsv
import numpy as np
```

```python
a = np.array([[4,2,1,-3],
              [1,2,-1,0],
              [3,-1,2,4],
              [0,2,4,3]],dtype=float)
```

```python
b1 = np.transpose(np.array([4,2,8,9],dtype=float))
```

```python
b2 = np.transpose(np.array([9,1,8,4],dtype=float))
```

```python
b3 = np.transpose(np.array([4,2,-7,5],dtype=float))
```

```python
lu, ov = gedo(a, pivot=False)
```

```python
print(lu)
```

```python
x1 = dosv(lu, ov, b1)
```

```python
print(x1)
```

```python
x2 = dosv(lu, ov, b2)
```

```python
print(x2)
```

```python
x3 = dosv(lu, ov, b3)
```

```python
print(x3)
```

```python
crlu, ov = gecr(a, pivot=False)
```

```python
print(crlu)
```

```python
x1 = crsv(crlu, ov, b1)
```

```python
print(x1)
```

```python
x2 = crsv(crlu, ov, b2)
```

```python
print(x2)
```

```python
x3 = crsv(crlu, ov, b3)
```

```python
print(x3)
```
