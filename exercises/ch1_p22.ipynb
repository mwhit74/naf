Each of the following has complex roots. Apply Newton's
method to find them<br>

x^2 = -2<br>
x^3 - x^2 - 1 = 0<br>
x^4 -3x^2 + x + 1 = 0<br>
x^2 = (e^(-2x) - 1)/x<br>

The estimated number and type of roots is based on Decartes' rule of signs. See Appendix A of the text book

```python
from naf.nonlin import newtone
import cmath
import numpy as np
import matplotlib.pyplot as plt
```

The possible roots of equation 1 are:<br>
2 or zero pos-real roots<br>
0 neg-real roots<br>
2 or zero imginary roots<br>

```python
f1 = lambda x: pow(x,2) + 2
df1 = lambda x: 2*x
```

The possible roots of equation 2 are:<br>
1 pos-real root<br>
0 neg-real roots<br>
2 imaginary roots<br>

```python
f2 = lambda x: pow(x,3) - pow(x,2) -1
df2 = lambda x: 3*pow(x,2) - 2*x
```

The possible roots of equation 3 are:<br>
2 or 0 pos-real roots<br>
2 or 0 neg-real roots<br>
4 or 2 or 0 imaginary roots<br>

```python
f3 = lambda x: pow(x,4) - 3*pow(x,2) + x + 1
df3 = lambda x: 4*pow(x,3) - 6*x + 1
```

The possible roots of equation 4 are:
2 or 0 pos-real roots<br>
0 neg-real roots<br>
2 or 0 imaginary roots<br>

```python
f4 = lambda x: (cmath.exp(-2*x)-1)/x - pow(x,2)
df4 = lambda x: -2*x - 2*cmath.exp(-2*x)/x - (-1 + cmath.exp(-2*x))/x**2
```

```python
c1 = 1 +1j
c2 = 1 -1j
c3 = 10 +5j
c4 = 10 -5j
c5 = c1
c6 = -1 -1j
c7 = c3
c8 = -10 -5j
```

```python
print('Eq. 1 roots')
n1 = newtone(f1, df1, c1)
print(n1)
n1 = newtone(f1, df1, c2)
print(n1)
```

```python
print('Eq. 2 roots')
n1 = newtone(f2, df2, c5)
print(n1)
n1 = newtone(f2, df2, c2)
print(n1)
n1 = newtone(f2, df2, c7)
print(n1)
```

```python
print('Eq. 3 roots')
n1 = newtone(f3, df3, c5)
print(n1)
n1 = newtone(f3, df3, c6)
print(n1)
n1 = newtone(f3, df3, c7)
print(n1)
n1 = newtone(f3, df3, c8)
print(n1)
```

```python
print('Eq. 4 roots')
n1 = newtone(f4, df4, c1)
print(n1)
n1 = newtone(f4, df4, c2)
print(n1)
```

```python
def graph():
    vf1 = np.vectorize(f1)
    vf2 = np.vectorize(f2)
    vf3 = np.vectorize(f3)
    vf4 = np.vectorize(f4)
    
    xr = np.linspace(-10,10)
    xr3b = np.linspace(0.5, 1.5)
    xr3a = np.linspace(-2.5, 2.5)
    xr4 = np.linspace(-3,5)
    
    plt.grid()
    plt.plot(xr, vf1(xr))
    plt.show()
    
    plt.grid()
    plt.plot(xr, vf2(xr))
    plt.show()
    
    plt.grid()
    plt.plot(xr3a, vf2(xr3a))
    plt.show()
    
    plt.grid()
    plt.plot(xr3a, vf3(xr3a))
    plt.show()
    
    plt.grid()
    plt.plot(xr3b, vf3(xr3b))
    plt.show()
    
    plt.grid()
    plt.plot(xr4, vf4(xr4))
    plt.show()

graph()
```
